```python
import pandas  as pd   # To load data
```

```python
cities_data= pd.read_excel('cities.xlsx') # load data set
```

```python
cities_data.head(10)  # Show 3 row of data set
```

```python
import folium  # To create map
```

```python
# To create Iran map, we use Latitude and longitude of Iran

iran_map=folium.Map(location = [32.4279,53.6880], zoom_start =5)  
iran_map
```

```python
# Illustrate city on the map

for lat, lng in zip(cities_data.latitude, cities_data.longitude):
    folium.CircleMarker([lat,lng], radious=2, color='green', fill=True, fill_color='red',
                        fill_opacity=0.5).add_to(iran_map)
iran_map
```

```python
# We want to show cities related to thier population

# We must create the map again

iran_map=folium.Map(location=[32.4279, 53.6880], zoom_start=5)

# To create cities:

cities_data['population_normalized']=((cities_data['population']-cities_data['population'].min())/cities_data['population'].max())*50

for row in range(cities_data.shape[0]):
    
    folium.CircleMarker(location=[cities_data.iloc[row,:]['latitude'],
                        cities_data.iloc[row,:]['longitude']],
                         radius=cities_data.iloc[row,:]['population_normalized'],
                        color='blue',
                        fill=True,
                        fill_color='red',
                        fill_opacity = 0.7).add_to(iran_map)
    
iran_map
```

```python
iran_map=folium.Map(location=[32.4279, 53.6880], zoom_start=5)

from folium.plugins import HeatMap

iran_map=folium.Map(location=[32.4279,53.6880], zoom_start=5)

iran_map

cities_data['population_normalized']=(cities_data['population']-cities_data['population'].min())/cities_data['population'].max()
HeatMap(data=list(zip(cities_data['latitude'],
                     cities_data['longitude'],
                     cities_data['population_normalized'])),
                    radius=20).add_to(iran_map)
iran_map
```
